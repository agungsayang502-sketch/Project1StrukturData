```python
# Program Pengolahan Nilai Mahasiswa

# 1. Input 10 nilai
nilai = []

for i in range(10):
    n = int(input(f"Masukkan nilai ke-{i+1}: "))
    nilai.append(n)

# 2. Nilai tertinggi dan terendah
nilai_tertinggi = max(nilai)
nilai_terendah = min(nilai)

# 3. Rata-rata
rata_rata = sum(nilai) / len(nilai)

# 4. Jumlah mahasiswa lulus (>= 60)
lulus = 0
for n in nilai:
    if n >= 60:
        lulus += 1

# 5. Tampilkan hasil
print("\n=== HASIL ===")
print("Nilai:", nilai)
print("Nilai tertinggi:", nilai_tertinggi)
print("Nilai terendah:", nilai_terendah)
print("Rata-rata:", rata_rata)
print("Jumlah lulus:", lulus)

# 6. Grafik sederhana
import matplotlib.pyplot as plt

# Grafik nilai
plt.figure()
plt.plot(nilai)
plt.title("Grafik Nilai Mahasiswa")
plt.xlabel("Mahasiswa ke-")
plt.ylabel("Nilai")
plt.show()

# Grafik kelulusan
kategori = ["Lulus", "Tidak Lulus"]
jumlah = [lulus, len(nilai) - lulus]

plt.figure()
plt.bar(kategori, jumlah)
plt.title("Grafik Kelulusan")
plt.show()
```
